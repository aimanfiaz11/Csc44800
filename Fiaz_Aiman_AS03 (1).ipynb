```python
#Aiman Fiaz
#Professor Erik
#Assignment 03

import time
import math

#naive summation
def naive_sum(N): 
    sum, value = 0, math.sqrt(23)
    for i in range(N):
        sum = sum + value
    return sum

#compensated summation
def comp_sum(N):
    s, value, e = 0, math.sqrt(23), 0
    for i in range(N):
        temp = s
        y = value + e
        s = temp + y
        e = (temp - s) + y
    return s + e

#different cases of N
for N in [1000, 10000, 100000, 1000000, 10000000, 100000000]: 
    exact_sum = N*math.sqrt(23) #exact value of the sum using different N cases
    print('N =', N, '\n')
    print('Exact Sum:', exact_sum, '\n')
    
#naive summation
    start_time = time.time() # start time
    naive = naive_sum(N) # compute naive sum
    end_time = time.time() # end time
    print('Naive Sum:', naive)
    print('Absolute Error:', abs(naive-exact_sum))#formula for absolute error
    print('Relative Error:', abs((naive-exact_sum)/exact_sum))#formula for relative error
    print('Elapsed Time:', (end_time-start_time), '\n')#formula for elapsed time
    
#compensated summation
    start_time = time.time() # start time
    comp = comp_sum(N) # compute  sum
    end_time = time.time() # end time
    print('Compensated Sum:', comp)
    print('Absolute Error:', abs(comp-exact_sum))#formula for absolute error
    print('Relative Error:', abs((comp-exact_sum)/exact_sum))#formula for relative error
    print('Elapsed Time:', (end_time-start_time))#formula for elapsed time
    print('\n')
```
